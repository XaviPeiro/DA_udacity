```python
import pandas as pd
import numpy as np
from plotly import __version__
print(__version__)
import cufflinks as cf
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
```

```python
init_notebook_mode(connected=True)
```

```python
cf.go_offline()
```

```python
df = pd.DataFrame(np.random.randn(100,5), columns='Q W E R T'.split())
```

```python
df
```

```python
df.iplot()
```

```python
df.iplot(kind='scatter', x='Q', y='W')
```

```python
df.iplot(kind='scatter', x='Q', y='W', mode='markers')
```

```python
df1 = pd.DataFrame({'cats': 'Q W E R'.split(), 'vals':[1,2,3,4]},)
df1
```

```python
df1.iplot(kind='bar', x='cats', y='vals')
```

```python
df.sum().iplot(kind='bar')
```

```python
df.median().iplot(kind='bar')
```

```python
df.iplot(kind='box')
```

```python
df2 = pd.DataFrame({'x': list(map(int,'11 22 33 44'.split())), 'y':[1,2,3,4], 'z': list(map(int, '12 41 14 53'.split()))})
df2
```

```python
df2.iplot(kind="box")
```

```python
df2.iplot(kind='surface', colorscale="rdylbu")
```
