```python
# Kernel density distribution ?
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set_style('whitegrid')
```

```python
tit = sns.load_dataset('titanic')
```

```python
tit.head()
```

```python
sns.jointplot(x='fare', y='age', data=tit)
```

```python
sns.distplot(tit["fare"], bins=30, kde=True, color='red')
```

```python
sns.boxplot(x="class", y='age', data=tit, palette="rainbow")
```

```python
sns.swarmplot(x="class", y='age', data=tit, palette="BuGn")
```

```python
sns.countplot(x="sex", data=tit)
```

```python
sns.heatmap(tit.corr(), cmap="coolwarm")
```

```python
grid = sns.FacetGrid(data=tit, col="sex")
grid.map(plt.hist, 'age')
```
