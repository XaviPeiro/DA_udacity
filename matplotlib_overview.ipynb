```python
import matplotlib.figure
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0, 100)
y = x*2
z = x**2
```

```python
fg = plt.figure()
fg.add_axes([0,0,1,1])
plt.ylabel("y_axis")
plt.xlabel("x_axis")
plt.title("title")
plt.plot(x,y)
```

```python
fg = plt.figure()

ax1 = fg.add_axes([0,0,1,1])
ax2 = fg.add_axes([.2,.5,.2,.2])
```

```python
ax1.set_xlabel("x")
ax1.set_ylabel("y")

ax2.set_xlabel("x")
ax2.set_ylabel("y")

ax1.plot(x,y)
ax2.plot(x,y)

fg
```

```python
fg = plt.figure()

ax1 = fg.add_axes([0,0,1,1])
ax2 = fg.add_axes([.2,.5,.4,.4])
```

```python
ax1.plot(x,z)
ax1.set_xlabel('X')
ax1.set_ylabel('Z')


ax2.plot(x,y)
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_title('zoom')
ax2.set_xlim(20,22)
ax2.set_ylim(30,50)

fg
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
```

```python
axes[0].plot(x,y,color="blue", lw=3, ls='--')
axes[1].plot(x,z,color="red", lw=5, ls='-')
fig
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2,figsize=(15,6))
ax1: matplotlib.figure.Axes
ax2: matplotlib.figure.Axes

ax1.plot(x,y,color="blue", lw=5)
ax1.set_ylabel('y',fontsize=20, color="green")
ax1.set_xlabel('x',fontsize=20, color="green")
ax1.tick_params(axis='x', colors='blue', labelsize=15.0)
ax1.tick_params(axis='y', colors='blue', labelsize=15.0)

ax2.plot(x,z,color="red", lw=3, ls='--')
ax2.tick_params(axis='x', colors='red', labelsize=15.0)
ax2.tick_params(axis='y', colors='red', labelsize=15.0)
ax2.set_ylabel('y',fontsize=20, color="green")
ax2.set_xlabel('x',fontsize=20, color="green")
```
